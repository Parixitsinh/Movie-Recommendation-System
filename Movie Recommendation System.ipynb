## 3] Taking relevant features for recommendation

```python
New_data = ['genres','keywords','tagline','cast','director']
New_data
```

```python
# Replacing the null values with null string

for feature in New_data:
    df[feature] = df[feature].fillna('')
```

```python
# Combining all the 5 selected features

df_new = df['genres']+' '+df['keywords']+' '+df['tagline']+' '+df['cast']+' '+df['director']
df_new
```

## 4] Getting the similarity score

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
vectorizer = TfidfVectorizer()
```

```python
Train_data = vectorizer.fit_transform(df_new)
```

```python
Similarity = cosine_similarity(Train_data)
Similarity
```

```python
Similarity.shape
```

```python
Movie = input('Enter movie name : ')
```

```python
# Creating a list with all the movie names given in the dataset

Movie_list = df['title'].tolist()
Movie_list
```

```python
import difflib
```

```python
# Finding the closest match for the movie name given as the input

Matching_names = difflib.get_close_matches(Movie, Movie_list)
Matching_names
```

```python
Best_match = Matching_names[0]
Best_match
```

```python
# Finding the index of this movie name

Index = df[df.title == Best_match]['index'].values[0]
Index
```

```python
Similarity_score = list(enumerate(Similarity[Index]))
Similarity_score
```

```python
len(Similarity_score)
```

```python
# Sorting the movies based on their similarity score

Sort_similar_movies = sorted(Similarity_score, key = lambda x:x[1], reverse = True)
Sort_similar_movies
```

## 5] Building a recommendation system

```python
# Print the name of similar movies based on the index

print('Movie suggestions for you : \n')

i = 1

for movie in Sort_similar_movies:
    index = movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if(i<15):
        print(i, '.',title_from_index)
        i+=1
```

```python
Movie = input('Enter movie name : ')

Movie_list = df['title'].tolist()

Matching_names = difflib.get_close_matches(Movie, Movie_list)

Best_match = Matching_names[0]

Index = df[df.title == Best_match]['index'].values[0]

Similarity_score = list(enumerate(Similarity[Index]))

Sort_similar_movies = sorted(Similarity_score, key = lambda x:x[1], reverse = True)

print('Movie suggestions for you : \n')

i = 1

for movie in Sort_similar_movies:
    index = movie[0]
    title_from_index = df[df.index==index]['title'].values[0]
    if(i<15):
        print(i, '.',title_from_index)
        i+=1
```
